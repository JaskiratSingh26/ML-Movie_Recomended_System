```python
import pickle

import pandas as pd
import numpy as np
import json

credit = pd.read_csv('tmdb_5000_credits.csv/tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv/tmdb_5000_movies.csv')

final_df = movies.merge(credit, on='title', how='left')

final_df.head(4)

# //stared data processsing || data cleaning

# checking 5 rows  of mering data
# print(final_df.head(5))

# checking colms name
# print(final_df.info())

# imp colms
# id,kwyword,lang,title,overview,geners,realese_date,spoken_languages,cast ,crew

final_df = final_df[['genres', 'id', 'spoken_languages', 'overview', 'title', 'cast', 'crew', 'movie_id','keywords']]


# print(final_df.head(2))
# checking the head again and checking if any data is missing or not
# final_df.info()
# final_df.isnull().sum()


# missing data is presetn in overview so we are removing it
#
# final_df.dropna(subset=['overview'], inplace=True)
# final_df.isnull().sum()


# checling is there any dupliate data
# final_df.duplicated()

#
# final_df['genres'][0]
# data = json.loads(final_df['genres'][0])
# for item in data:
#     print(item['name'])


def helper_genres(obj):
    genres_list = []
    data = json.loads(obj)
    for item in data:
        genres_list.append(item['name'])

    return genres_list


final_df['genres'] = final_df['genres'].apply(helper_genres)

final_df.head(2)



```

```python



def helper_lang(obj):
    langs_list = []
    for item in json.loads(obj):
        langs_list.append(item['name'])

    return langs_list


# helper_lang([{"iso_639_1": "en", "name": "English"}, {"iso_639_1": "es", "name": "Espa\u00f1ol"}])
final_df['spoken_languages'] = final_df['spoken_languages'].apply(helper_lang)

final_df.head(5)




```

```python

# Now, define the helper function
def helper_cast(obj):
    cast_list = []  # Initialize an empty list
    counter = 0  # Counter to limit entries

    for item in json.loads(obj):  # Iterate over the list of dictionaries
        if counter < 3:  # Ensure only 3 items are added
            cast_list.append(item['name'])
            counter += 1
        else:
            break  # Exit loop when counter reaches 3

    return cast_list# Return the final list

# Apply the function
final_df['cast'] = final_df['cast'].apply(helper_cast)

# Check results
final_df.head(3)
```

```python


def finding_director(obj):
    names=[]
    for item in json.loads(obj):
       if item['job'] == 'Director':
           names.append(item['name'])

    return names


final_df['crew']=final_df['crew'].apply(finding_director)
final_df.head(3)
```

```python
final_df['cast']=final_df['cast'].apply(lambda x: [name.replace(" ", "") for name in x])
final_df['genres']=final_df['genres'].apply(lambda x: [name.replace(" ", "") for name in x])
final_df['crew']=final_df['crew'].apply(lambda x: [name.replace(" ", "") for name in x])
final_df.head(5)
#
# final_df['cast'][0].replace(" ", "")
#
# final_df['cast'] = final_df['cast'].apply(lambda x: "".join(x.split()))
# print(final_df['cast'])


# print(final_df['cast'].apply(lambda x: [name.replace(" ", "") for name in x])




```

```python
def keywords_name(obj):
    names=[]
    for item in json.loads(obj):
        names.append(item['name'])

    return names



final_df['keywords']=final_df['keywords'].apply(keywords_name)
final_df['keywords']=final_df['keywords'].apply(lambda x:[name.replace(" ", "") for name in x] )




final_df.head(5)
```

```python
final_df['Tags'] = final_df['genres'] + final_df['keywords'] + final_df['cast'] + final_df['crew'] + final_df['overview'].apply(lambda x: [x])
final_df.head(3)
```

```python
working_dataframe=final_df[['id','Tags','spoken_languages','movie_id','title']].copy()

working_dataframe['Tags'] = working_dataframe['Tags'].apply(lambda x: " ".join(map(str, x)) if isinstance(x, list) else str(x))
working_dataframe['Tags'] = working_dataframe['Tags'].apply(lambda x: x.lower())

working_dataframe.head(2)
```

```python
working_dataframe.head(6)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import nltk
cv=CountVectorizer(max_features=5000,stop_words='english' )
vector=cv.fit_transform(working_dataframe['Tags']).toarray()

selected_words = cv.get_feature_names_out()
len(selected_words)
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

ps = PorterStemmer()

def stemming(text):
    whole_row=[]
    for word in text.split():
        whole_row.append(ps.stem(word))
    return ' '.join(whole_row)





working_dataframe["Tags"]=working_dataframe['Tags'].apply(stemming)
working_dataframe.head(2)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
def recommendations(movie_name):
    try:
        movie_index = working_dataframe[working_dataframe['title'] == movie_name].index[0]
    except IndexError:
        print(f"Movie '{movie_name}' not found in dataset.")
        return

    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:11]

    print(f"\nTop recommendations for '{movie_name}':")
    for value in movie_list:
        print(f"- {value[0]} and this the movie{working_dataframe.iloc[value[0]].title}")



recommendations("Batman Begins")




# working_dataframe.iloc[8].title



```
